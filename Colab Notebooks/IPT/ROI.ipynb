```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from skimage import io, color, filters, img_as_float, img_as_ubyte

# Load the image and convert to grayscale
img = io.imread('snake.jpeg')
if len(img.shape) == 3:
    img = color.rgb2gray(img)
img = img_as_float(img)


# Compute global threshold using Otsu's method
threshold = filters.threshold_otsu(img)

# Apply global thresholding to create binary mask
binary = img > threshold

# Apply ROI mask to the original image
roi = img * binary

original = img

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(original, cmap='gray')
axes[0].set_title('Original Image')
axes[0].axis('off')

axes[1].imshow(binary, cmap='gray')
axes[1].set_title('Binary Mask (ROI)')
axes[1].axis('off')

axes[2].imshow(roi, cmap='gray')
axes[2].set_title('Region of Interest')
axes[2].axis('off')

plt.tight_layout()
plt.show()




```
