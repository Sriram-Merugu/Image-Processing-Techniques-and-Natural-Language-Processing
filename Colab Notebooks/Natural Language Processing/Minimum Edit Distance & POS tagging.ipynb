```python
def minDistance(word1: str, word2: str) -> int:
  m = len(word1)
  n = len(word2)
  dp = [[0]*(n+1) for _ in range(m+1)]
  for i in range(1, m + 1):
    dp[i][0] = i
  for j in range(1, n + 1):
    dp[0][j] = j
  cost = 1 # to perform any operation cost is 1
  for i in range(1, m + 1):
    for j in range(1, n + 1):
      if word1[i-1] == word2[j-1]:
        dp[i][j] = dp[i-1][j-1]
      else:
        dp[i][j] = min(dp[i-1][j], dp[i][j-1], dp[i-1][j-1]) + cost
  print("DP Table is:\n")
  for i in dp:
    print(i)
  return dp[m][n]
word1="horse"
word2="rose"
distance=minDistance(word1,word2)
print(distance)
```
