```python
import nltk
from nltk.corpus import movie_reviews
from nltk.classify import NaiveBayesClassifier
from nltk.tokenize import word_tokenize
```

```python
def extract_features(words):
    return dict([(word, True) for word in words])
```

```python
nltk.download('movie_reviews')
nltk.download('punkt')

positive_reviews = []
negative_reviews = []

for field in movie_reviews.fileids('pos'):
    words = word_tokenize(movie_reviews.raw(field))
    positive_reviews.append((extract_features(words), 'positive'))

for field in movie_reviews.fileids('neg'):
    words = word_tokenize(movie_reviews.raw(field))
    negative_reviews.append((extract_features(words), 'negative'))


positive_cutoff = int(len(positive_reviews) * 0.8)
negative_cutoff = int(len(negative_reviews) * 0.8)

train_reviews = positive_reviews[:positive_cutoff] + negative_reviews[:negative_cutoff]
test_reviews = positive_reviews[positive_cutoff:] + negative_reviews[negative_cutoff:]

classifier = NaiveBayesClassifier.train(train_reviews)
accuracy = nltk.classify.util.accuracy(classifier, test_reviews)
print("Accuracy:", accuracy)
```

```python
def analyze_sentiment(text):
    words = word_tokenize(text)
    features = extract_features(words)
    sentiment = classifier.classify(features)
    return sentiment

test_sentences = [
    'i will win',
    'this is not a good movie'
]

for sentence in test_sentences:
    sentiment = analyze_sentiment(sentence)
    print(f"Sentence: {sentence}, Sentiment: {sentiment}")
```
