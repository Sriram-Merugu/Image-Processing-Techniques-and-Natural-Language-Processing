```python
# Text and Word normalization

str = "Hey! This is a Good Movie. Now, I am watching"
data = str.lower()
data
```

```python
import string
print(string.punctuation)

def punctuation(text):
  punctuationfree="".join([i for i in text if i not in string.punctuation])
  return punctuationfree

#storing the puntuation free text
free_punc= punctuation(data)
free_punc

```

```python
import nltk
nltk.download('punkt')
```

```python
from nltk.tokenize import word_tokenize
def tokenization(text):
  tokens=word_tokenize(text)
  return tokens

tkn = tokenization(free_punc)
tkn
```

```python
nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('english')

```

```python
def remove_stopwords(text):
  return [i for i in text if i not in stopwords]

tkn_without_stopwords = remove_stopwords(tkn)
tkn_without_stopwords
```

```python
import pandas as pd
from nltk.stem.porter import PorterStemmer

porter_stemmer = PorterStemmer()
stem_word = " ".join([porter_stemmer.stem(word) for word in tkn_without_stopwords])
stem_word
```

```python
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
lemm_word = " ".join([wordnet_lemmatizer.lemmatize(word) for word in tkn_without_stopwords])
lemm_word
```

```python
# For telugu

string = "ప్ర ప్ంచంలో భారత్ ఒక అమూల్య రత్న ం"
# string = punctuation(string)
tkn = word_tokenize(string)

stopwords_telugu = ['గురంచి', 'పై', 'నిజంగా', 'త్ర్వా త్', 'మళ్ళీ', 'కాదు', 'అందర', 'అందరూ', 'దాదాపు',
'మాత్ర మే', 'వంట', 'ఇప్ప టికే', 'కూడా', 'అయితే', 'ఎపుప డు', 'వదద', 'మధ్య', 'ఒక', 'మరయు', 'మరొక', 'ఏ', 'ఎవరోఒకరు', 'ఎవరై నా', 'ఏదై నా', 'ఎకక డై నా', 'వేరుగా', 'మెచ్చు కో', 'త్గిన', 'ఉనాన రు', 'కాదు', 'చ్చట్టూ ']


def stopword(text):
  sw = ' '.join([i for i in text if i not in stopwords_telugu])
  return sw
stopword_free_telugu_2 = stopword(tkn)
print("Stopword-free Text:", stopword_free_telugu_2)
```
