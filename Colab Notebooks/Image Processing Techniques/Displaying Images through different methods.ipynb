```python
# reading image from local
from skimage import io
img = io.imread('download.jpeg')
plt.imshow(img)
plt.axis('off')
```

```python
# convert img to grey scale
img2 = io.imread('download.jpeg', as_gray=True)
plt.imshow(img2, cmap='gray')
```

```python
# convert img values from float to int
from skimage import img_as_ubyte

img3 = img_as_ubyte(io.imread('download.jpeg', as_gray=True))
print(img3.size * img3.itemsize) # image size in bytes
```

```python
# Gray
image = 0
if len(img.shape) == 3:
  image = np.mean(img, axis=2)
  # plt.imshow(image)

# Negate the image
neg_image = 255 - image
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(neg_image, cmap='gray')
plt.axis('off')

# plt.show()
```

```python
# Threshold
# convert gray scale to BW by applying threshold
threshold = 10
bw_img = (image > threshold)*255

# Display B/W image
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')

plt.subplot(1, 2, 2)
plt.imshow(bw_img, cmap='gray')
```
