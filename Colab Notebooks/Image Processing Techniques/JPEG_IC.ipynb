```python
# JPEG Compression technique
import imageio.v2 as imageio
import os
import matplotlib.pyplot as plt

# Read the image
image_path = 'snake.jpeg'
image = imageio.imread(image_path)

# Define compression level (0-100, higher is better quality but larger size)
compression_level = 100

# Compress and save the image
compressed_image_path = 'c_snake.jpeg'
imageio.imwrite(compressed_image_path, image, quality=compression_level)


# Get the size of original and compressed images
original_size = os.path.getsize(image_path)
compressed_size = os.path.getsize(compressed_image_path)

# Print sizes
print(f"Original image size: {original_size / 1024:.2f} KB")
print(f"Compressed image size: {compressed_size / 1024:.2f} KB")

# Display input and compressed images
plt.figure(figsize=(12, 6))
# Input image
plt.subplot(1,2,1)
plt.imshow(image)
plt.title('Input Image')
plt.axis('off')

plt.subplot(1,2,2)
# Compressed image
compressed_image = imageio.imread(compressed_image_path)
plt.imshow(compressed_image)
plt.title(f'Compressed Image (Quality={compression_level})')
plt.axis('off')
```
