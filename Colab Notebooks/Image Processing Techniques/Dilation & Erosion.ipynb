```python
import numpy as np
from scipy import ndimage
import matplotlib.pyplot as plt
from skimage import io, color, morphology, img_as_float, img_as_ubyte

img = io.imread('snake.jpeg')
if len(img.shape) == 3:
  img = color.rgb2gray(img)
img = img_as_float(img)

# Binarize
binary = img > 0.5

kernel_size = 3
kernel = morphology.square(kernel_size)
# Dilation
plt.subplot(1,2,1)
plt.title('Dilation')
dilated = morphology.dilation(binary, kernel)
plt.imshow(dilated)

# Erosion
plt.subplot(1,2,2)
plt.title("Erosion")
eroded = morphology.erosion(binary, kernel)
plt.imshow(eroded)

```
