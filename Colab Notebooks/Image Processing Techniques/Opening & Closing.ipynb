```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from skimage import io, color, morphology, img_as_float, img_as_ubyte

img = io.imread('snake.jpeg')
if len(img.shape) == 3:
  img = color.rgb2gray(img)
img = img_as_float(img)

# Binarize the image

binary_image = img > 0.5

# Apply opening
kernel_size = 3
kernel = morphology.square(kernel_size)
# erosion followed by dilation(opening)
opened = ndimage.binary_opening(binary_image, kernel)

plt.subplot(1,2,1)
plt.title("Opening")
plt.imshow(opened)


# Apply closing
closed = ndimage.binary_closing(binary_image, kernel)

plt.subplot(1, 2, 2)
plt.title("Closing")
plt.imshow(closed)
```
