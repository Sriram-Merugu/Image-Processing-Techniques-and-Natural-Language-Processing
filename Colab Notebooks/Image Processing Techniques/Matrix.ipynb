```python
mat1 = np.array([[1,2,3], [4,5,6]])
mat2 = np.asmatrix('[1 2;3 4]')
```

```python
#dense matrix
print(mat1, mat2)
import scipy

# sparse matrix
sm = scipy.sparse.csr_matrix(mat2)
print(sm)
```

```python
from scipy.linalg import eig
eigval, eigvec = eig(mat2)
print(eigval, eigvec)
```

```python
#matrix decompositions
from scipy.linalg import svd, lu, qr

U, S, V = svd(mat2)
P, L, U = lu(mat2)
Q, R = qr(mat2)
```

```python
# Norms

frobenius_norm = np.linalg.norm(mat2)
L1_norm = np.linalg.norm(mat2, ord=1)
L2_norm = np.linalg.norm(mat2, ord=2)
infinity_norm = np.linalg.norm(mat2, ord=np.inf)
```
